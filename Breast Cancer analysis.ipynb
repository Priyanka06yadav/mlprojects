```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df =pd.read_csv('breast cancer data.csv') 
```

```python
df.replace({'M':1,'B':0},inplace =True)
```

```python
columns = ['id','Unnamed: 32']
df =df.drop(columns,axis=1)
```

```python
df.describe()
```

```python
sns.heatmap(df.corr()>0.9,annot =False)
plt.show()
```

```python
count =1
fig = plt.figure(figsize=(15,10))
cols = df.columns.tolist()
cols.remove('diagnosis')

for col in cols:
    plt.subplot(6, 5, count)
    sns.boxplot(x = df[col])
    count += 1
plt.tight_layout()
plt.show()
```

```python
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3-Q1

    lower =Q1-1.5*IQR
    upper = Q3+1.5*IQR
    df.loc[df[col]>=upper,col]=upper
    df.loc[df[col]<=lower,col]=lower
```

```python
X = df.iloc[:,1:31].values
```

```python
kmeans = KMeans(n_clusters =2,init='k-means++',n_init=10,max_iter=300,tol=0.0001,verbose=0,random_state=42)
Y_predict = kmeans.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X, Y_predict, metric='euclidean')
# Print the score
print('Silhouette Score: %.3f' % score)
```

```python
u_labels = np.unique(Y_predict)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(X[Y_predict == i][:,0] , X[Y_predict == i][:, 1], label =i)
plt.legend()
plt.show()
```

```python
kmeans.inertia_
```

```python
SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(X)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on = 'Country Code', how ='left')
final_df.head()
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_value = final_df.Country.value_counts().values
```

```python
# Pie chart -Top 3 countries only
plt.pie(country_value[:3],labels= country_name[:3] , autopct='%1.2f%%')
```

Observations :
Zomato maximum records or transactions are from India the from USA and then UK

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating count'})
```

```python
ratings
```

#### Observation

     1.When ratings are between 4.5 to 4.9 -----Excellent
     2.When ratings are between 4.0 to 4.4 -----Ver Good
     3.When ratings are between 3.5 to 3.9 -----Good
     4.When ratings are between 2.5 to 3.4 -----Average
     5.When ratings are between 1.8 to 2.4 -----Poor
     6.When ratings are 0                  -----Not rated

```python
plt.figure(figsize=(25,6))
sns.barplot('Aggregate rating', 'Rating count', data = ratings)
```

```python
plt.figure(figsize=(25,6))
sns.barplot('Aggregate rating', 'Rating count',hue ='Rating color' ,data = ratings ,palette =['blue','red','orange','yellow','green','green'])
```

#### Observations
1. Not rated count is high
2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x= 'Rating color', data = ratings,palette =['blue','red','orange','yellow','green','green'])
```

```python
### Find the countries name that has given 0 rating
final_df[final_df['Aggregate rating'] ==0].groupby('Country').size().reset_index()
```

#### Observation
Maximum number of 0 rating are from Indian customers
